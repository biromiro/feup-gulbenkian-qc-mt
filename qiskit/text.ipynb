```python
%matplotlib inline

from qiskit import *
from qiskit.visualization import *
from qiskit.tools.monitor import *
```

```python
circ = QuantumCircuit(2,2) # We use one qubit and also one classical bit for the measure result 

circ.h(0) #We apply the H gate
circ.h(1)

circ.barrier()
circ.measure(1,1) # We measure
circ.barrier()

circ.h(0)
circ.h(1)
circ.barrier()

circ.measure(range(2),range(2)) # We measure

circ.draw(output='mpl', style={'displaycolor': {'h': '#289B92'}}) #We draw the circuit
```

```python
backend_sim = Aer.get_backend('qasm_simulator') # We choose the backend

job_sim = execute(circ, backend_sim, shots=1024) # We execute the circuit, selecting the number of repetitions or 'shots'

result_sim = job_sim.result() # We collect the results

counts = result_sim.get_counts(circ) # We obtain the frequency of each result and we show them 
print(counts) 
plot_histogram(counts, color='#289B92')
```
